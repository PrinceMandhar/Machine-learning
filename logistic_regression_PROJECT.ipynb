```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# Create the dataset
x, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)

# Plot the data points
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='winter', s=100)
plt.colorbar()  # To show the color scale
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Synthetic Classification Data')
plt.show()
```

```python
# def perceptron(x,y):
#     x = np.insert(x,0,1,axis=1)
#     weight = np.ones(x.shape[1])
#     lr = 0.1 
#     for i in range(1000):
#         j=np.random.randint(0,100)
#         y_hat = step(np.dot(x[j],weight))
#         weight = weights + lr*(y[j]-y_hat)*x[j]

#     return weight[0],weight[1: ]


import numpy as np

def perceptron(x, y, lr=0.01, epochs=1000):
    # Initialize weights
    weights = np.zeros(x.shape[1] + 1)  # +1 for the bias term

    # Add bias term to input features
    x = np.hstack([np.ones((x.shape[0], 1)), x])

    # Training loop
    for epoch in range(epochs):
        j = np.random.randint(0, len(y))
        y_hat = step(np.dot(x[j], weights))
        weights += lr * (y[j] - y_hat) * x[j]

    return weights[0], weights[1:]
```
